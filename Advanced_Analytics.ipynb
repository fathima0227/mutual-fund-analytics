```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# 1. GENERATE / LOAD MUTUAL FUND DATA
def load_mutual_fund_data():
    print("--- 1. Loading Mutual Fund Portfolio Dataset ---")
    # Setting a random seed for reproducible financial simulation
    np.random.seed(42)

    # Simulating data resembling actual AMFI Indian mutual fund schemes
    funds = ['SBI Bluechip Fund', 'HDFC Mid-Cap Opportunities', 'Axis Small Cap Fund',
             'ICICI Prudential Prudential Equity & Debt', 'Parag Parikh Flexi Cap',
             'UTI Nifty 50 Index Fund', 'Mirae Asset Large Cap', 'Nippon India Small Cap']

    data = pd.DataFrame({
        'Fund_Name': funds,
        'Category': ['Large Cap', 'Mid Cap', 'Small Cap', 'Hybrid', 'Flexi Cap', 'Index', 'Large Cap', 'Small Cap'],
        'Initial_NAV': [100.0, 85.5, 45.0, 120.0, 50.0, 95.0, 110.0, 35.0],
        'Current_NAV': [155.2, 148.3, 92.1, 178.4, 112.5, 144.2, 162.8, 79.6],
        'Years_Invested': [3, 3, 3, 3, 3, 3, 3, 3],
        'Risk_Free_Rate_%': [6.0, 6.0, 6.0, 6.0, 6.0, 6.0, 6.0, 6.0] # standard Indian RBI repo/T-Bill approximations
    })

    # Adding simulated historical standard deviations (annualized volatility)
    data['Annual_Volatility_%'] = [12.5, 16.8, 22.1, 9.8, 14.2, 11.1, 12.1, 24.5]
    print(data.head(3))
    return data

# 2. FINANCIAL ADVANCED ANALYTICS (CAGR & SHARPE RATIO)
def calculate_financial_metrics(df):
    print("\n--- 2. Computing Advanced Financial Performance Metrics ---")

    # CAGR Formula: ((Current / Initial) ** (1 / Years)) - 1
    df['CAGR_%'] = (((df['Current_NAV'] / df['Initial_NAV']) ** (1 / df['Years_Invested'])) - 1) * 100

    # Sharpe Ratio Formula: (Asset Return - Risk Free Return) / Volatility
    df['Sharpe_Ratio'] = (df['CAGR_%'] - df['Risk_Free_Rate_%']) / df['Annual_Volatility_%']

    for idx, row in df.iterrows():
        print(f"{row['Fund_Name']} ({row['Category']}): CAGR = {row['CAGR_%']:.2f}%, Sharpe Ratio = {row['Sharpe_Ratio']:.2f}")
    return df

# 3. MACHINE LEARNING: RISK-RETURN PERFORMANCE CLUSTERING
def perform_fund_clustering(df):
    print("\n--- 3. Executing K-Means Machine Learning Clustering ---")

    # Extracting core performance features for segmentation
    features = ['CAGR_%', 'Annual_Volatility_%', 'Sharpe_Ratio']
    X = df[features]

    # Scaling metrics for even distribution weighting
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Applying unsupervised K-Means to bucket funds into 3 tiers (e.g., High Return/High Risk, Conservative, Balanced)
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    df['Performance_Cluster'] = kmeans.fit_transform(X_scaled).argmax(axis=1)

    print("\nMutual Fund Clusters successfully generated:")
    print(df[['Fund_Name', 'Category', 'CAGR_%', 'Annual_Volatility_%', 'Performance_Cluster']])
    return df

# 4. PREDICTIVE MODELING: TREND FORECASTING
def predict_nav_growth(df):
    print("\n--- 4. Running Linear Regression Trend Analysis ---")

    # Simulating standard continuous timeline asset mapping
    timeline_years = np.array([0, 1, 2, 3]).reshape(-1, 1)

    # Evaluating a sample targeted fund (e.g., SBI Bluechip Fund growth slope)
    sbi_nav_history = np.array([100.0, 118.0, 134.5, 155.2])

    model = LinearRegression()
    model.fit(timeline_years, sbi_nav_history)

    # Forecast forward to year 4 and year 5
    future_years = np.array([4, 5]).reshape(-1, 1)
    predictions = model.predict(future_years)

    print(f"Projected NAV for SBI Bluechip Fund in Year 4: {predictions[0]:.2f} INR")
    print(f"Projected NAV for SBI Bluechip Fund in Year 5: {predictions[1]:.2f} INR")

# MAIN PIPELINE EXECUTION
if __name__ == "__main__":
    fund_data = load_mutual_fund_data()
    analyzed_data = calculate_financial_metrics(fund_data)
    clustered_data = perform_fund_clustering(analyzed_data)
    predict_nav_growth(clustered_data)
    print("\n--- Advanced Analytics Pipeline Completed Successfully! ---")
```
